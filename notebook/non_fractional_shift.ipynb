```python
import numpy as np
import matplotlib.pyplot as plt
import math
from numpy import fft
```

```python
n = 50
x0 = n * 0.5
x = np.arange(n)
```

```python
z = x < x0
plt.xlim(x0 - 20, x0 + 20)
plt.stem(x,z, basefmt="gray")
```

```python
delta = 10
H = np.exp(-2j * x/ n * np.pi * delta)
```

```python
z_shifted = fft.ifft(H * fft.fft(z))
```

```python
plt.xlim(x0 - 20, x0 + 20)
plt.stem(x, np.real(z_shifted))
```
